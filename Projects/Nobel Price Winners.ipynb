The Nobel Prize has been among the most prestigious international awards since 1901. Each year, awards are bestowed in chemistry, literature, physics, physiology or medicine, economics, and peace. In addition to the honor, prestige, and substantial prize money, the recipient also gets a gold medal with an image of Alfred Nobel (1833 - 1896), who established the prize.

![](Nobel_Prize.png)

The Nobel Foundation has made a dataset available of all prize winners from the outset of the awards from 1901 to 2023. The dataset used in this project is from the Nobel Prize API and is available in the `nobel.csv` file in the `data` folder.

In this project, you'll get a chance to explore and answer several questions related to this prizewinning data. And we encourage you then to explore further questions that you're interested in!

```python
# Loading in required libraries
import pandas as pd
import seaborn as sns
import numpy as np

# Start coding here!
df = pd.read_csv('data/nobel.csv')
```

```python
df.head()
```

```python
# What is the most commonly awarded gender and birth country?
# Gender:
genders_count = df['sex'].value_counts().sort_values(ascending = False)
top_gender = genders_count.index[0]
print(f'Top gender is: {top_gender}')
# Country:
countries_count = df['birth_country'].value_counts().sort_values(ascending = False)
top_country = countries_count.index[0]
print(f'Top country is: {top_country}')
```

```python
# Which decade had the highest ratio of US-born Nobel Prize winners to total winners in all categories?
df['decade'] = (df['year'])//10*10
decade_national = df.groupby(['decade', 'birth_country']).size()
decade_national_total = decade_national.groupby('decade').sum()
us_nobels_per_decade = decade_national.loc[(slice(None),'United States of America')]
ratio = us_nobels_per_decade/decade_national_total
max_decade_usa = ratio.idxmax()
print(f'Decade whith highest us-total ratio is: {max_decade_usa}')
```

```python
# Which decade and Nobel Prize category combination had the highest proportion of female laureates?
df_tot = df.groupby(['decade', 'category']).size().reset_index()
df_fem = df[df['sex'] == 'Female'].groupby(['decade', 'category']).size().reset_index()
df_merged = pd.merge(df_tot, df_fem, on=['decade', 'category'], how='left')
df_merged.columns = ['decade', 'category', 'total_winners', 'female_winners']
df_merged['female_winners'] = df_merged['female_winners'].fillna(0)
df_merged['female_proportion'] = df_merged['female_winners'] / df_merged['total_winners']
max_proportion = df_merged['female_proportion'].max()
max_decade = df_merged[df_merged['female_proportion'] == max_proportion]
max_female_dict = {max_decade['decade'].values[0]: max_decade['category'].values[0]}
print('The decade-category with max female ratio is')
print(max_female_dict)
```

```python
# Who was the first woman to receive a Nobel Prize, and in what category?
first_woman =  df[df['sex'] == 'Female'].iloc[0]
first_woman_name = first_woman['full_name']
first_woman_category = first_woman['category']
print(first_woman_name)
print(first_woman_category)
```
